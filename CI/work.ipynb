```python

# Scale numeric features
scaler = StandardScaler()
yeast_features = scaler.fit_transform(yeast.iloc[:, :-1])
liver_features = scaler.fit_transform(liver.iloc[:, :-1])

# Separate target
yeast_target = yeast['localization_site']
liver_target = liver['Selector']   

yeast_features
```

```python
# after scaling data, lets split the data into training and testing sets

from sklearn.model_selection import train_test_split

# use yeast_x_train 
yeast_x_train, yeast_x_test, yeast_y_train, yeast_y_test = train_test_split(yeast_features, yeast_target, test_size=0.2, random_state=42)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import time

# Separate features and target for both datasets
yeast_features = yeast.drop(columns=["localization_site"])
yeast_target = yeast["localization_site"]

liver_features = liver.drop(columns=["Selector"])
liver_target = liver["Selector"]

# Replace liver target values to binary (0, 1)
liver_target = liver_target.replace({2: 0, 1: 1})

# Split datasets into train and test sets
yeast_x_train, yeast_x_test, yeast_y_train, yeast_y_test = train_test_split(
    yeast_features, yeast_target, test_size=0.2, random_state=42
)

liver_x_train, liver_x_test, liver_y_train, liver_y_test = train_test_split(
    liver_features, liver_target, test_size=0.2, random_state=42
)

# Define the XGBoost model with binary classification configuration
xgb_model = XGBClassifier(
    use_label_encoder=False,  # Suppresses warning for label encoding
    eval_metric='logloss',    # For binary classification (use logloss for logistic regression)
    objective='binary:logistic', # Binary classification objective
    max_depth=5,              # Maximum tree depth
    n_estimators=100,         # Number of trees
    learning_rate=0.1,        # Learning rate
    random_state=42           # Seed for reproducibility
)

# Function to train, evaluate, and measure execution time
def train_and_evaluate(x_train, x_test, y_train, y_test, dataset_name):
    """
    Trains the XGBoost model, evaluates it, and measures execution time.
    """
    print(f"\n--- Evaluating XGBoost on {dataset_name} Dataset ---")
    
    # Train the model
    start_time = time.time()
    xgb_model.fit(x_train, y_train)
    end_time = time.time()
    
    # Make predictions
    predictions = xgb_model.predict(x_test)
    
    # Evaluate model
    report = classification_report(y_test, predictions)
    execution_time = end_time - start_time
    
    # Print results
    print(report)
    print(f"Execution Time: {execution_time:.4f} seconds")
    return report, execution_time

# Evaluate on Yeast dataset
yeast_report, yeast_time = train_and_evaluate(
    yeast_x_train, yeast_x_test, yeast_y_train, yeast_y_test, "Yeast"
)

# Evaluate on Liver dataset
liver_report, liver_time = train_and_evaluate(
    liver_x_train, liver_x_test, liver_y_train, liver_y_test, "Indian Liver"
)

# Summary of results
print("\n--- Summary of Results ---")
print(f"Yeast Dataset: Execution Time: {yeast_time:.4f} seconds")
print(f"Liver Dataset: Execution Time: {liver_time:.4f} seconds")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
import time

# Define the Random Forest model
rf_model = RandomForestClassifier(
    n_estimators=100,  # Number of trees
    max_depth=5,        # Maximum depth of the tree
    random_state=42     # Seed for reproducibility
)

# Function to train, evaluate, and measure execution time
def train_and_evaluate_rf(x_train, x_test, y_train, y_test, dataset_name):
    print(f"\n--- Evaluating Random Forest on {dataset_name} Dataset ---")
    
    # Train the model
    start_time = time.time()
    rf_model.fit(x_train, y_train)
    end_time = time.time()
    
    # Make predictions
    predictions = rf_model.predict(x_test)
    
    # Evaluate model
    report = classification_report(y_test, predictions)
    execution_time = end_time - start_time
    
    # Print results
    print(report)
    print(f"Execution Time: {execution_time:.4f} seconds")
    return report, execution_time

# Evaluate on Liver dataset using Random Forest
liver_rf_report, liver_rf_time = train_and_evaluate_rf(
    liver_x_train, liver_x_test, liver_y_train, liver_y_test, "Indian Liver"
)

# Summary for Random Forest
print(f"Liver Dataset (Random Forest): Execution Time: {liver_rf_time:.4f} seconds")
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
import time

# Define the AdaBoost model
ada_model = AdaBoostClassifier(
    n_estimators=100,  # Number of boosting rounds
    random_state=42     # Seed for reproducibility
)

# Function to train, evaluate, and measure execution time
def train_and_evaluate_ada(x_train, x_test, y_train, y_test, dataset_name):
    print(f"\n--- Evaluating AdaBoost on {dataset_name} Dataset ---")
    
    # Train the model
    start_time = time.time()
    ada_model.fit(x_train, y_train)
    end_time = time.time()
    
    # Make predictions
    predictions = ada_model.predict(x_test)
    
    # Evaluate model
    report = classification_report(y_test, predictions)
    execution_time = end_time - start_time
    
    # Print results
    print(report)
    print(f"Execution Time: {execution_time:.4f} seconds")
    return report, execution_time

# Evaluate on Liver dataset using AdaBoost
liver_ada_report, liver_ada_time = train_and_evaluate_ada(
    liver_x_train, liver_x_test, liver_y_train, liver_y_test, "Indian Liver"
)

# Summary for AdaBoost
print(f"Liver Dataset (AdaBoost): Execution Time: {liver_ada_time:.4f} seconds")
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
import time

# Define the SVM model
svm_model = SVC(
    kernel='linear',     # Linear kernel for binary classification
    C=1,                 # Regularization parameter
    random_state=42      # Seed for reproducibility
)

# Function to train, evaluate, and measure execution time
def train_and_evaluate_svm(x_train, x_test, y_train, y_test, dataset_name):
    print(f"\n--- Evaluating SVM on {dataset_name} Dataset ---")
    
    # Train the model
    start_time = time.time()
    svm_model.fit(x_train, y_train)
    end_time = time.time()
    
    # Make predictions
    predictions = svm_model.predict(x_test)
    
    # Evaluate model
    report = classification_report(y_test, predictions)
    execution_time = end_time - start_time
    
    # Print results
    print(report)
    print(f"Execution Time: {execution_time:.4f} seconds")
    return report, execution_time

# Evaluate on Liver dataset using SVM
liver_svm_report, liver_svm_time = train_and_evaluate_svm(
    liver_x_train, liver_x_test, liver_y_train, liver_y_test, "Indian Liver"
)

# Summary for SVM
print(f"Liver Dataset (SVM): Execution Time: {liver_svm_time:.4f} seconds")
```

```python
from sklearn.metrics import classification_report
import numpy as np

# Function to get average precision, recall, and F1 scores
def get_average_scores(y_test, predictions):
    report = classification_report(y_test, predictions, output_dict=True)
    avg_precision = report['weighted avg']['precision']
    avg_recall = report['weighted avg']['recall']
    avg_f1_score = report['weighted avg']['f1-score']
    return avg_precision, avg_recall, avg_f1_score

# For Random Forest
rf_avg_precision, rf_avg_recall, rf_avg_f1 = get_average_scores(liver_y_test, rf_model.predict(liver_x_test))

# For AdaBoost
ada_avg_precision, ada_avg_recall, ada_avg_f1 = get_average_scores(liver_y_test, ada_model.predict(liver_x_test))

# For SVM
svm_avg_precision, svm_avg_recall, svm_avg_f1 = get_average_scores(liver_y_test, svm_model.predict(liver_x_test))

# Print average results
print(f"Random Forest - Avg Precision: {rf_avg_precision:.4f}, Avg Recall: {rf_avg_recall:.4f}, Avg F1 Score: {rf_avg_f1:.4f}")
print(f"AdaBoost - Avg Precision: {ada_avg_precision:.4f}, Avg Recall: {ada_avg_recall:.4f}, Avg F1 Score: {ada_avg_f1:.4f}")
print(f"SVM - Avg Precision: {svm_avg_precision:.4f}, Avg Recall: {svm_avg_recall:.4f}, Avg F1 Score: {svm_avg_f1:.4f}")
```
